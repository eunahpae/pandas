```python
# map({함수명},{자료형데이터})

def change(x):
    result = x ** 2
    return result

_list = [1,2,3,4]
#  _List의 원소의 값들을 제곱 [1,4,9,16]

# 새로운 비어있는 리스트를 생성
result = []

# 반복문을 이용하여 함수를 호출
for i in _list:
    result.append(change(i))

result
```

```python
list(map(change,_list))
```

```python
# apply({함수명})
# 스리즈 형태의 데이터를 일반 함수에 대입하여 새로운 스리즈 형태의 데이터를 생성

# 일반함수
def change(x):
    # x가 0이라면 '월'
    if x ==0:
        result = '월'
    elif x ==1:
        result = '화'
    elif x ==2:
        result = '수'
    elif x ==3:
        result = '목'
    elif x ==4:
        result = '금'
    elif x ==5:
        result = '토'
    else:
        result = '일'
    return result
    
```

```python
df['요일'].apply(change)
```

```python
# lamda 함수
# 함수를 선언함과 동시에 호출
_list = [1,3,6,8]

list(
    map(
        lambda x : x ** 2,
        _list
    )
)
```

```python
def change2(x):
    # x를 위치값으로 생각하여 자료형 데이터에서 해당하는 위치의 데이터를 출력
    week_list = ['월','화','수','목','금','토','일']
    return week_list[x]

df['요일'] = df['요일'].apply(change2)
```

```python
# 데이터를 그룹화하여 연산
# 그룹화 함수
# groupby({특정한 컬럼}).연산함수
# 연산함수 : 최소, 최대, 합계, 평균, 표준편차, 개수
df[['요일','일일사망자','일일확진자']].groupby(['요일']).sum()
```

```python
group_data = df[['요일','일일사망자','일일확진자']].groupby(['요일']).agg(['mean','sum'])
```

```python
# 한글깨짐 방지
# matplot에 있는 폰트의 설정을 변경

# OS가 mac인 경우
if platform.system() == 'Darwin':
    plt.rc('font',family='AppleGothic')
else : 
    plt.rc('font',family='Malgun Gothic')
```

```python
# 바형 그래프
# bar({x측데이터},{y축데이터}, option)
# 요일별 일일확진자의 평균 값을 바형 그래프 표기 
x = group_data.index
y = group_data['일일확진자']['mean']

plt.bar(x,y)
plt.show()
```

```python
# 요일별로 데이터프레임의 순서 변경
week_order = ['월','화','수','목','금','토','일']
group_data= group_data.loc[week_order]
```

```python
# 일일확진자읭 ㅛ일별 평균값 그래프와
# 일일사망자의 요일별 합계값 그래프를
# subplot을 이용하여 시각화

x = group_data.index
y1 = group_data['일일확진자']['mean']
y2 = group_data['일일사망자']['sum']

# 그래프가 그려지는 영역의 사이즈를 지정
plt.figure(figsize=(14,8))

# 영역을 2분할하여 첫번째 영역 선택
plt.subplot(2,1,1)
# 그래프 시각화
plt.bar(x,y1, color=['blue','green'])
# x 축의 이름
plt.xlabel('요일')
plt.ylabel('평균')
# 그래프의 제목
plt.title('일일확진자 평균')


# 2개의 영역중 두번째 영역 선택
plt.subplot(2,1,2)
plt.bar(x,y2)
plt.show()
```
